### Converting Categorical Variables into Numerical Variables
- Here is only one categorical variable which is Sex, we can convert it into numerical variables using Label Encoding, 
- target variable is Rings, we can use it as a numerical variable

```python
df["Sex"].value_counts()
```

### Gonna use label encoder method to convert it into numerical variables

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["Sex"] = label_encoder.fit_transform(df["Sex"])
```

```python
df
```

### check labelEncoder mapped the categories using `label_encoder.classes_` to see the original categories

```python
# Create a new encoder with original data
new_label_encoder = LabelEncoder()
original_categories = ['F', 'I', 'M']
new_label_encoder.fit(original_categories)
print("Original categories:", new_label_encoder.classes_)
print("\nMapping:")
for i, category in enumerate(new_label_encoder.classes_):
    print(f"{category} -> {i}")
```

## Now let's separate the features and target variable

```python
df.columns
```

```python
 # x = df.drop(columns=["Rings"]) 
x = df.drop("Rings", axis=1)
y = df["Rings"]
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

## Let's Train the model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

# Test Model Performance on test set

```python
lr.score(x_test, y_test)
```

### Now let's predict the age of the abalone using the coefficients and intercept

```python
lr.coef_
```

```python
x_test.iloc[1]["Height"]
```

```python
df.columns
```

```python
sex = x_test.iloc[1]["Sex"]
length = x_test.iloc[1]["Length"]
diameter = x_test.iloc[1]["Diameter"]
height = x_test.iloc[1]["Height"]
whole_weight = x_test.iloc[1]["Whole weight"]
shucked_weight = x_test.iloc[1]["Shucked weight"]
viscera_weight = x_test.iloc[1]["Viscera weight"]
shell_weight = x_test.iloc[1]["Shell weight"]

```

```python
predicted_age = (
  lr.coef_[0] * sex + lr.coef_[1] * length + lr.coef_[2] * diameter + lr.coef_[3] * height + lr.coef_[4] * whole_weight + lr.coef_[5] * shucked_weight + lr.coef_[6] * viscera_weight + lr.coef_[7] * shell_weight + lr.intercept_
)
```
